```python
import sqlite3
from dotenv import load_dotenv
import os

_ = load_dotenv()
DB_PATH = os.getenv('DB_PATH')
```

```python
sql_statement = '''
    select * from inc
'''
```

```python
with sqlite3.connect(DB_PATH) as conn:
    cur = conn.cursor()
    cur.execute(sql_statement)
    data = cur.fetchall()

print(data)
```

```python
# Empty inc table

with sqlite3.connect(DB_PATH) as conn:
    cur = conn.cursor()
    cur.execute('''
        CREATE TABLE inc_new AS
        SELECT * FROM inc WHERE 1=0
    ''')
    conn.execute('DROP TABLE inc')
    conn.execute('ALTER TABLE inc_new RENAME TO inc')
    conn.commit()
```
